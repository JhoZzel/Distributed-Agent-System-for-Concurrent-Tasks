```python
import numpy as np
from skimage.io import imread, imsave
import matplotlib.pyplot as plt
```

```python
im_roja  = imread('main/img_roja.jpg')
im_verde = imread('main/img_verde.jpg')
im_azul = imread('main/img_azul.jpg')
```

```python
alto, ancho, capas = im_roja.shape
print("El tamaño de la imagen es: ")
print("Alto: ", alto, "px\nAncho:", ancho, "px\nCapas:", capas, "px")
```

```python
# mostrando la imagen con filtro rojo...
plt.imshow(im_roja)
```

```python
# mostrando la capa 0 de la imagen con filtro rojo
plt.imshow(im_roja[:,:,0], cmap='gray')
```

```python
# mostrando la capa 1 y 2 de la imagen con filtro rojo...
plt.imshow(im_roja[:,:,1], cmap='gray')
```

```python
plt.imshow(im_roja[:,:,2], cmap='gray')
```

```python
# mostrando la imagen con filtro verde...
plt.imshow(im_verde)
```

```python
# tomamos las partes de cada imagen filtrada recordando el orden RGB
parte_roja  = im_roja[:,:,0]  # R
parte_verde = im_verde[:,:,1] # G
parte_azul  = im_azul[:,:,2]  # B
```

```python
imagen_completa = np.stack([parte_roja, parte_verde, parte_azul], axis=-1)
```

```python
# mostremos la imagen completa...
plt.imshow(imagen_completa)
```

```python
# Pregunta: ¿Qué pasaría si se tomara otras partes de las imágenes filtradas?
# Ejemplo: parte_roja  = im_roja[:,:,1]
# ¿Daría como resultado la misma imagen?

parte_roja  = im_roja[:,:,1]  # R
parte_verde = im_verde[:,:,2] # G
parte_azul  = im_azul[:,:,0]  # B
imagen_completa = np.stack([parte_roja, parte_verde, parte_azul], axis=-1)
plt.imshow(imagen_completa)
```

```python
n = 7
fig, axs = plt.subplots(n, 1, figsize=(5, 35), dpi=100)

for i in range(n):
    img_roja  = imread(f"img/{i + 1}R.jpg")
    img_verde = imread(f"img/{i + 1}G.jpg")
    img_azul = imread(f"img/{i + 1}B.jpg")
    parte_roja  = img_roja[:,:,0]  # R
    parte_verde = img_verde[:,:,1] # G
    parte_azul  = img_azul[:,:,2]  # B
    imagen_completa = np.stack([parte_roja, parte_verde, parte_azul], axis=-1)
    
    axs[i].imshow(imagen_completa)
    
plt.subplots_adjust(hspace=0.01)
plt.show()
```
